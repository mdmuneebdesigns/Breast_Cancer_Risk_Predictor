```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
df.diagnosis.unique()
```

```python
if df.isnull().sum().sum()>0:
    for column in df.columns:
        if df.column.dtype != 'object':
            df.column = df.column.fillna(df.column.median())
        else:
            df[column] = df.columns.fillna(df[column].mode()[0])
```

```python
# Agar missing values hain toh median se impute karein
if df.isnull().sum().sum() > 0:
    for column in df.columns:
        if df[column].dtype != 'object':
            df[column] = df[column].fillna(df[column].median())
        else:
            df[column] = df[column].fillna(df[column].mode()[0])
```

```python
X = df.drop("diagnosis" , axis=1)
y = df.diagnosis
```

```python
pd.DataFrame(y)
```

```python
from sklearn.model_selection import train_test_split
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25 , random_state=42 , stratify=y)
print(f"""       
****** Traning set size : {X_train.shape} ******
***** Test set size  : {X_test.shape} *******
""")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

# Pipelines create karna
models = {
    'Logistic Regression': Pipeline([('scaler', StandardScaler()), ('classifier', LogisticRegression(max_iter=1000))]),
    'KNN': Pipeline([('scaler', StandardScaler()), ('classifier', KNeighborsClassifier())]),
    'SVM': Pipeline([('scaler', StandardScaler()), ('classifier', SVC(kernel='rbf', probability=True))]),
    'Decision Tree': Pipeline([('classifier', DecisionTreeClassifier())]),  # Decision tree ko scaling ki zaroorat nahi
    'XGBoost': Pipeline([('classifier', XGBClassifier(eval_metric='logloss', use_label_encoder=False))])
}

# Train aur evaluate karna
for name, pipeline in models.items():
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{name}: {accuracy:.4f}")
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid = {
    'classifier__C': [0.1, 1, 10, 100],
    'classifier__gamma': [1, 0.1, 0.01, 0.001],
    'classifier__kernel': ['rbf']
}

# Create pipeline
svm_pipeline = Pipeline([('scaler', StandardScaler()), ('classifier', SVC())])

# GridSearchCV
grid = GridSearchCV(svm_pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid.fit(X_train, y_train)

print("Best params:", grid.best_params_)
print("Best CV accuracy:", grid.best_score_)

# Test on unseen data
y_pred = grid.best_estimator_.predict(X_test)
print("Test accuracy:", accuracy_score(y_test, y_pred))
```
